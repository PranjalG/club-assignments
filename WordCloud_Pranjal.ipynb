Hints: 
1.   split() function for splitting the words
2.   isalpha() for detecting a non-alphabetical character
3.   the format of dictionary: keys = word , value = frequency of the word



```python
def calculate_frequencies(file_content):
   
    # TYPE YOUR MAIN CODE HERE
    frequencies = {}
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she","not", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more","for","there","some", "such", "no", "nor", "too", "very", "can", "will", "just"]
    
    my_list = file_content.split()
    words = []
    for i in my_list:
        if ((len(i) > 2) & (i.isalpha() == True)):
            i = i.lower()
            if i not in uninteresting_words:
                words.append(i)
        else:
            continue
    
    for i in words:
        if(i in frequencies):
            frequencies[i] += 1
        else:
            frequencies[i] = 1
    
    my_dict = frequencies
    #wordcloud
    cloud = wordcloud.WordCloud(width=2160, height=1080, background_color = "white")
    cloud.generate_from_frequencies(my_dict)
    return cloud.to_array()
```

```python
# Display your wordcloud image

path = "D:\Gitanjali.txt"                        #copy the path of the file here
file1 = open(path, 'r')
file_contents = file1.read()
myimage = calculate_frequencies(file_contents)
plt.figure(figsize=(20,10))
plt.imshow(myimage)
plt.axis('off')
name = "Queen"                                  # TYPE YOUR NAME HERE
plt.savefig("{}.jpeg".format(name))
plt.show()
```
